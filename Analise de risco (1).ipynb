```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ds= pd.read_csv('planilharisco.csv' , delimiter=';', encoding='latin-1')
```

```python
ds
```

```python
def riscoTipoEmprego(tipo):
    if tipo =='publico':
        return  2 *3  #pontuação x peso
    if tipo =='clt':
        return  3 *3
    if tipo =='Contrato 1 ano':
        return  5 *3
    else:
        return  5 *3
    
```

```python
def riscoTempoEmpresa(tempo):
    if tempo == 0 or tempo == 1:
        return 5 *2
    if tempo == 2 or tempo == 3:
        return 4 *2
    if tempo == 4 or tempo == 5:
        return 3 *2
    if tempo == 6 or tempo == 7:
        return 2 *2
    if tempo > 7:
        return 1 *2
    else:
        return 5 *2   
   
```

```python
def riscoQtdEmprestimo(qtd):
    if qtd == 0 or qtd == 1:
        return 5 *2 
    if qtd == 2 or qtd == 3:
        return 4 *2 
    if qtd == 4 or qtd == 5:
        return 3 *2 
    if qtd >= 6:
        return 2 *2 
    else:
        return  5 *2   
    
```

```python
ds['Risco_TipoEmprego'] = ds['tipo emprego'].apply(riscoTipoEmprego)
```

```python
ds.head(5)
```

```python
ds['Risco_Tempo_Empresa'] = ds['tempo empresa'].apply(riscoTempoEmpresa)
```

```python
ds
```

```python
ds['Risco_Qtd_emprestimo'] = ds['numero de emprestimos feitos'].apply(riscoQtdEmprestimo)
```

```python
ds
```

```python
ds['Risco_'] = (ds['Risco_TipoEmprego']+ ds['Risco_Tempo_Empresa']+ ds['Risco_Qtd_emprestimo'])/3
ds
```

```python
ds['Risco_']=  np.around(ds['Risco_'], decimals=3)
```

```python
#ordenando pelo risco
ds.sort_values(by='Risco_',ascending=False)
```
